1. Bu dersimizde görmediğimiz fakat çok yaygın olarak kullanılan çizge türlerinden birisi de çubuk grafikleri (bar plot)'dir. Çubuk grafikleri özellikle, kategorik (yani sınırlı sayıda farklı değer içeren) değişkenleri incelemek için yararlı araçlardır. Bu ödevde, kategorik bir değişken içeren bir veri kümesi bulmanız gerekecek. Bunun için Kaggle sitesindeki veri setlerine bir göz atabilirsiniz. Daha sonra, bulduğunuz veri kümesini Pandas'ı kullanarak bir veri çerçevesine aktarın. Veri kümenizde yer alan kategorik değişkenlerden iki tanesini seçin ve subplot kullanarak, bu iki değişkenin çubuk grafiklerini çıkarın. Bulgularınızı mentörünüzle tartışmayı unutmayın!

```python
import pandas as pd
```

```python
# Veri 1982 ve 2017 yillarinda Teksas'ta idam edilen 543 mahkumdan kaydedilmis ilk 10 veri uzerine calisilmistir. 
TexasLastStatement = pd.read_csv('TexasLastStatement.csv')
TexasLastStatement.head(10)
```

```python
TexasLastStatement2 = TexasLastStatement.copy()
TexasLastStatement2[['Race', 'CountyOfConviction']].head()
```

```python
import seaborn as sns

plt.figure(figsize=(60,20))
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 23 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 30 }

plt.subplot(1,2,1)
sns.barplot(TexasLastStatement2["Race"], TexasLastStatement2["Age"])
plt.title("Race & Age", fontdict=baslik_font)
plt.xlabel('Race', fontdict=eksen_font )
plt.ylabel('Age', fontdict=eksen_font )


plt.subplot(1,2,2)
sns.barplot(TexasLastStatement2["CountyOfConviction"], TexasLastStatement2["Age"])
plt.title("CountyOfConviction& Age", fontdict=baslik_font)
plt.xlabel('CountyOfConviction', fontdict=eksen_font )
plt.ylabel('Age', fontdict=eksen_font )
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(50, 10))
ax = TexasLastStatement2.plot.bar(x='Race', y='EducationLevel', colormap='Paired')
plt.show()
```

```python
plt.figure(figsize=(8, 20))
plt.bar(TexasLastStatement2.Race, TexasLastStatement2.CountyOfConviction)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(TexasLastStatement2['Age'], color='purple')
plt.ylabel('Yas Degerleri')
plt.title('Rastgele Diziler')

plt.subplot(1, 2, 2)
plt.plot(TexasLastStatement2['EducationLevel'], color='green')
plt.ylabel('Egitim Seviyesi')
plt.title('Ötelenmiş Diziler')

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(TexasLastStatement2['Age'], color='purple')
plt.ylabel('age')
plt.title('Yas Dagilimi')

plt.subplot(1, 2, 2)
plt.scatter(x = TexasLastStatement2['EducationLevel'], y = TexasLastStatement2['Age'], color='green')
plt.ylabel('Egitim')
plt.title('Egitim')

plt.tight_layout()
plt.show()
```
